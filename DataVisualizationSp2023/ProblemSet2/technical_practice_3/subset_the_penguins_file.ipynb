```python
# Import pandas to work with the data
import pandas as pd

# Read the penguins file
penguins = pd.read_csv("penguins.csv")
penguins
```

```python
# Create a new data frame with the rows with any missing data dropped
penguins_subsetted = penguins.dropna(axis = 0, 
                                     how ='any')
penguins_subsetted
```

```python
# Save the subsetted data as a file
penguins_subsetted.to_csv("penguins_subsetted.csv")
```
